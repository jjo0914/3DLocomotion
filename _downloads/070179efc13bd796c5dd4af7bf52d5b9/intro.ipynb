```python
# For tips on running notebooks in Google Colab, see
# https://docs.pytorch.org/tutorials/beginner/colab
%matplotlib inline
from google.colab import drive  # google.colab.drive를 drive.으로 사
drive.mount('/content/drive') # 경로고정/ 이후에는 /content/drive/MyDrive/ drive 내에 경로지정
# path :/content/drive/MyDrive/)
!pip install bvh
from bvh import Bvh # bvh. 를 Bvh로사용가능
path="/content/drive/MyDrive/dataset-1_walk-left_active_001.bvh"

with open(path,"r") as f: # with 문법 열고 자동닫기 or 끄고 다시켜기 등등
  data=Bvh(f.read()) # f.read로 전체문자열을 가져오고 bvh가 자동으로 파싱해준다
print("프레임 수:", data.nframes)
print("프레임 시간:", data.frame_time)
print("관절 이름:", data.get_joints_names())
print("UpperLeg_L offset :",data.joint_offset("UpperLeg_L"))
print("LowerLeg_L offset :",data.joint_offset("LowerLeg_L"))
print("Foot_L offset :",data.joint_offset("Foot_L"))
print("Toes_L offset :",data.joint_offset("Toes_L"))
print("UpperLeg_R offset :",data.joint_offset("UpperLeg_R"))
print("LowerLeg_R offset :",data.joint_offset("LowerLeg_R"))
print("Foot_R offset :",data.joint_offset("Foot_R"))
print("Toes_R offset :",data.joint_offset("Toes_R"))

print("Hipes offset :",data.joint_offset("Hips"))

##
joint_name = "Toes_L"
frame_idx = 0

channels = data.joint_channels(joint_name)

for ch in channels:
    value = data.frame_joint_channel(frame_idx, joint_name, ch)
    print(ch, value)

```
